```python
1. What is the role of the 'else' block in a try-except statement? Provide an example
scenario where it would be useful.
Answer:
The else block in a try-except statement is executed if no exception occurs in the try block.
This can be useful for executing code that should only be executed if no error occurs.
try:
  number = int(input("Enter a number: "))
except ValueError:
  print("That was not a valid number.")
else:
  print("You entered the number:", number)
```

```python
2. Can a try-except block be nested inside another try-except block? Explain with an
example.
Answer:
Yes, we can have nested try-except blocks in Python.  Can have try block inside another try block.
try:
  try:
    x = int(input("Enter a number: "))
  except ValueError:
    print("That was not a valid number.")
  else:
    try:
      y = x / 0
    except ZeroDivisionError:
      print("Division by zero error.")
    else:
      print("x / y = ", x / y)
except Exception as e:
  print("An error occurred:", e)
```

```python
3. How can you create a custom exception class in Python? Provide an example that
demonstrates its usage.
Answer:
We can create a custom exception class by defining a new class that inherits from the built-in Exception class or one of its subclasses.
This custom exception class should typically have an informative name that describes the specific error or exceptional condition we want to handle in your code.
we can also include additional attributes or methods in custom exception class to provide more context or functionality.
# Define a custom exception class
class CustomError(Exception):
    def __init__(self, message):
        super().__init__(message)

# Example usage of the custom exception
def divide(x, y):
    if y == 0:
        raise CustomError("Division by zero is not allowed")
    return x / y

# Try dividing two numbers, catching the custom exception
try:
    result = divide(10, 0)
except CustomError as e:
    print(f"Custom Exception: {e}")
else:
    print(f"Result: {result}")

```

```python
8. How can you set up logging to capture log messages from multiple modules or
classes in a Python application?
Answer:
We can use folling steps
Create a root logger. The root logger is the top-level logger in the logging hierarchy.
Create a logger for each module or class that you want to log messages from.
Set the level of each logger. The level of a logger determines the severity of the messages that are logged.
Add a handler to each logger. A handler is used to send log messages to a destination, such as a file or the console.
Use the logging.getLogger() function to get a logger object.

import logging

# Create a root logger
root_logger = logging.getLogger()

# Set the level of the root logger
root_logger.setLevel(logging.DEBUG)

# Create a logger for the my_module module
my_module_logger = logging.getLogger("my_module")

# Set the level of the my_module logger
my_module_logger.setLevel(logging.INFO)

# Create a handler for the my_module logger
file_handler = logging.FileHandler("my_module.log")
file_handler.setFormatter(logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s'))
my_module_logger.addHandler(file_handler)

# Log a message from the my_module module
my_module_logger.info("This is an info message from my_module")
```

```python
10. Write a Python program that logs a message to a file named "app.log" with the
following requirements:
● The log message should be "Hello, World!"
● The log level should be set to "INFO."
● The log file should append new log entries without overwriting previous ones.
Answer:
import logging

# Create a logger
logger = logging.getLogger()

# Set the level of the logger
logger.setLevel(logging.INFO)

# Create a file handler
file_handler = logging.FileHandler("app.log", mode="a")

# Set the formatter for the file handler
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
file_handler.setFormatter(formatter)

# Add the file handler to the logger
logger.addHandler(file_handler)

# Log a message
logger.info("Hello, World!")
```

```python
11. Create a Python program that logs an error message to the console and a file named
"errors.log" if an exception occurs during the program's execution. The error
message should include the exception type and a timestamp.
Answer:

import logging

# Create a logger
logger = logging.getLogger()

# Set the level of the logger
logger.setLevel(logging.ERROR)

# Create a file handler
file_handler = logging.FileHandler("errors.log")

# Set the formatter for the file handler
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
file_handler.setFormatter(formatter)

# Add the file handler to the logger
logger.addHandler(file_handler)

try:
  # Do something that might raise an exception
  raise ValueError("This is an error")
except Exception as e:
  # Log the error message
  logger.error(e)

```
